# Binary tree-Insert

```python
class BinHeap:
    
    def __init__(self):
        
        self.heapList = [0]       # 初始BinHeap => 1st格是0
        
        self.currentSize = 0      # 一開始currentSize是0
         
    def percUp(self,i):
        
        while i // 2 > 0:                                 # i就是currentSize       //2是確認currentSize是不是大於1，currentSize == 1時，不用檢測heap property
            
            if self.heapList[i] < self.heapList[i // 2]:  #因為是binary tree，所以只要位置//2，也就是sequence//2就代表是上一level (parent node)
                                                          #此處是看parent node是不是 > child node (如果是，就要modify，因為是min-heap)
                tmp = self.heapList[i // 2]               #先記錄parent node的值
                
                self.heapList[i // 2] = self.heapList[i]  #把parent node的值換成child node的值
                
                self.heapList[i] = tmp                    #把child node的值換成原parent node的值
                
            i = i // 2                                    # i // 2 => 繼續再往上一level檢查heap property，因為我們剛剛有做一次swap，所以要確認再上一level是否符合heap property

    def insert(self,k):                           #用insert來增加binary tree的elements
        
        self.heapList.append(k)                   #用append來加入element => appending guarantees that we'll maintain the complete shape property
                                                  
        self.currentSize = self.currentSize + 1   # currentsize + 1
        
        self.percUp(self.currentSize)             #因為有加入element，apeending likely violate the heap property => 所以要確認 heap property
                                                   
     
r = BinHeap()

r.insert(5)
r.insert(9)
r.insert(11)
r.insert(14)
r.insert(7)
```

# Binary tree-Extract

```python
class BinHeap:
    
    def __init__(self):
        
        self.heapList = [0]           
        self.currentSize = 0
        
    def percDown(self,i):
        
        while (i * 2) <= self.currentSize:             #第一run 的i == 1     i * 2 == 2 => 因為如果currentSize == 1時，不用檢測heap property
            
            mc = self.minChild(i)                      #找i == 1的original最小child node
            
            if self.heapList[i] > self.heapList[mc]:   #有i == 1的original最小child node後，如果此新node 1 > i == 1的original最小child node => 要modify (因為是min-heap)
                
                tmp = self.heapList[i]                 #先記錄node 1的值
                
                self.heapList[i] = self.heapList[mc]   #把node 1的值換成i == 1的original最小child node的值
                
                self.heapList[mc] = tmp                #把i == 1的original最小child node的值換成新node 1的值
                
            i = mc                                     # i == mc => 繼續再往下一level檢查heap property，因為我們剛剛有做一次swap，所以要確認再下一level是否符合heap property                        

    def minChild(self,i):
        
        if i * 2 + 1 > self.currentSize:                    #第一run 的i == 1    i * 2 + 1 == 3 => 因為如果currentSize == 2時，i == 1的original最小child node就是node 2
            
            return i * 2
        
        else:
            
            if self.heapList[i*2] < self.heapList[i*2+1]:   #第一run 的i == 1    i * 2 + 1 == 3 => 因為如果currentSize > 2時，i == 1的original最小child node就要用比較   
                                                            # 比較 node 2    &    node 3 哪一個比較小
                return i * 2
            
            else:
                
                return i * 2 + 1

    def delMin(self):
        
        retval = self.heapList[1]                            #紀錄list[1]的值，因為binary tree的extraction是從list[1]開始刪
        
        self.heapList[1] = self.heapList[self.currentSize]   #把list[1]的值換成最後一個element的值
        
        self.currentSize = self.currentSize - 1              # currentsize - 1
        
        self.heapList.pop()                                  # pop掉最後一個element，因為上述steps已經把list[1]的值換成最後一個element的值
        
        self.percDown(1)                                     #因為把list[1]的值換成最後一個element的值，所以要確認 heap property     
                                                             # i == 1   => 從1開始檢查
        return retval
```

# Binary tree-Building a heap with existing list

```python
class BinHeap:
    
    def __init__(self):
        
        self.heapList = [0]
        self.currentSize = 0

    def percDown(self,i):
        
        while (i * 2) <= self.currentSize:             #第一run 的i == 倒數第二level的parent node => 此處的判斷式一定會成立，因為原本是用len(alist) // 2得到i
            
            mc = self.minChild(i)                      #找i == 倒數第二level的parent node的最小child node
            
            if self.heapList[i] > self.heapList[mc]:   #有i == 倒數第二level的parent node的最小child node後，如果此parent node > 最小child node => 要modify (因為是min-heap)
                
                tmp = self.heapList[i]                 #先記錄original parent node的值
                
                self.heapList[i] = self.heapList[mc]   #把original parent node的值換成最小child node的值
                
                self.heapList[mc] = tmp                #把最小child node的值換成原original parent node的值
                
            i = mc                                     # i == mc => 繼續再往下一level檢查heap property，因為我們剛剛有做一次swap，所以要確認再下一level是否符合heap property                        

    def minChild(self,i):
        
        if i * 2 + 1 > self.currentSize:                    #第一run 的i == 倒數第二level的parent node => 因為如果currentSize == i * 2時，最小child node就是node (i * 2)
            
            return i * 2
        
        else:
            
            if self.heapList[i*2] < self.heapList[i*2+1]:   #第一run 的i == 倒數第二level的parent node => 因為如果currentSize > i * 2 時，最小child node就要用比較   
                                                            # 比較 node (i * 2)   &    node (i * 2 + 1) 哪一個比較小
                return i * 2  
            
            else:
                
                return i * 2 + 1

    def buildHeap(self,alist):
        
        i = len(alist) // 2                 #此處的len(alist) // 2的意思是從倒數第二level的node (parent node)開始檢查 (因為//2 & binary的關係，所以會找到倒數第二level)
                                            #因為最後一level就是leaf nodes                                    
        self.currentSize = len(alist)       #紀錄currentSize
        
        self.heapList = [0] + alist[:]      #加0，因為所有的binary tree的list都會從0開始
        
        while (i > 0):                      #從倒數第二level開始檢查直到root                
            
            self.percDown(i)                #用percDown() =>  從倒數第二level看最後一level        
            
            i = i - 1                       # i = i-1 => 從倒數第三level看倒數第二level，依序往上至root
                                            # (而在倒數第三level看倒數第二level時，可能有swap，所以還要再從倒數第二level看最後一level => 就是i == mc那line)
r = BinHeap()

r.buildHeap([9, 6, 5, 2, 3])
```
